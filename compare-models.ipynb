```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt

        
        
data=pd.read_csv('/kaggle/input/results/results.csv')

#create orders for the first two plots
mean_order = data.groupby('id_alg')['mean_ind'].sum().sort_values(ascending=False).index.values

std_order = data.groupby('id_alg')['std_ind'].sum().sort_values(ascending=True).index.values


# take the mean of ind and order them descendingly in order to see the alg with the highest mean
sns.barplot(x = "id_alg", y = "mean_ind", data = data, order=mean_order)
sns.set(rc={'figure.figsize':(16,16)})
plt.show()

# take the std of ind and order them ascendingly in order to see the alg with the lowest std
sns.barplot(x = "id_alg", y = "std_ind", data = data, order=std_order)
sns.set(rc={'figure.figsize':(15,15)})
plt.show()

# i've also tried to see the mean and std grouped by algorithm using hues but
# that looked really messy as you can see in the 3rd and 4th plot

#mean
sns.barplot(x = "id_alg", y = "mean_ind", hue = "id_dataset", data = data)
sns.set(rc={'figure.figsize':(15,15)})
plt.show()

#std
sns.barplot(x = "id_alg", y = "std_ind", hue = "id_dataset", data = data)
sns.set(rc={'figure.figsize':(15,15)})
plt.show()


# i've also tried the other way around, grouping them by dataset
# it looks colorful and nice, but no meaning can be extracted easly out of it
# i've also tried ordering these two but did not manage

sns.barplot(x = "id_dataset", y = "mean_ind", hue = "id_alg", data = data)
sns.set(rc={'figure.figsize':(15,15)})
plt.show()


sns.barplot(x = "id_dataset", y = "std_ind", hue = "id_alg", data = data)
sns.set(rc={'figure.figsize':(15,15)})
plt.show()




# as for conclusions:
# the algorithm with the highest mean was 10, followed by 12
# but 10 had an average std_dev
# and the one with the lowest std_dev was 8, followed by 7 and 12
# but 7 had an average mean
# so since 12 was in both these measurements in top 3,
# we can conclude that 12 was the most performant model
# with the 8th model as the second most performant

```
